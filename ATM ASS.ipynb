## Exercise 1

develop a zero-dimension Energy Balance Model, version1 (EBM0.1) by solving the following equation.
                         
                         S(1 − α) = 4εσ T^4


where s = solar energy (also known as solar constant, usually S = 1372 W m2)


$\alpha$  is the reflectivity of the Earth system (called planetary albedo; it ranges from 0 to 1; default
value = 0.3); ε represents the infrared transmissivity (ε = 0.61); σ is the Stefan-Boltzmann
constant (σ = 5.67 × $10^{-8} W m^{-2} K^{-4}$), which relates temperature to radiant emission; and T
is the global mean surface temperature of the earth system.

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Constant

s = 1372 #solar constant
a = 0.3 #𝛼  albedo
e = 0.61 #ε  emissitivity
c = 5.67 * 10**-8  #stefan-Boltzmann constant
```

```python
#a zero-dimension Energy Balance Model

T=(s*(1-a)/(4*e*c))**(0.25)

print("the global mean surface temperature of the earth system  = ",T-273.15 ,"c")
```

(1) develop a zero-dimension Energy Balance model, version 2 (EBM0.2) by solving the following equation.
$$ Cp\frac{∂T}{∂t}= S(1 − α) − 4εσT^4$$


where Cp is the specific heat capacity of the earth system (Cp = $1.05x10^{23} JK^{-1}$); t is time;
dt is the time 1.0x$10^{21}$ For all the experiments, integrate the model for 400 timesteps.


```python
cp = 1.05*10**(23)

Tn=273.15
dt=1.0*10**21
######
t=[]
temp=[]
time=0
for i in range(400):
    time=time+dt
    F= (s*(1-a)-(4*c*e*Tn**4))/cp
    Tf=Tn+F*dt
    t.append(time)
    temp.append(Tn)
    Tn=Tf

plt.plot(t,np.array(temp)-273.15)
plt.xlabel("Time $")
plt.ylabel("earth surface temperature (T) $c^0$")
```

### Exercise 3

 develop one-dimensional
energy balance model (EBM1) by solving the following equation

$$ Cp\frac{∂T_{lat}}{∂t}= S(1 − α) − 4εσT_{lat}^4 - k(T_{lat} - T_{global})$$


```python
#---- Set default constants
sigma    =  5.67e-8         # Stefan-Boltman constant
ems      =  0.6             # Emmisitivity
tc       =  3.81            # Transport constants
thc      =  1.05e23         # Total Heat Capacity
dt       =  1.0e21          # dt
nstep    =  400             # nstep
SR       =  1370            # Solar constant

sfrac  = np.zeros(18)
albedo = np.zeros(18)
Tlat   = np.zeros(18)

# Read inputs from file zone.dat
file1 = open("zone.dat","r")
file1.readline()

for i in range(18):
    zone,sfrac[i],albedo[i],Tlat[i] = file1.readline().split()
    Tlat[i]  = Tlat[i] + 273.14
file1.close()

# This function calculates T global
Tgg = 0.0
def Calculate_Tglobal(Temp,Tgg,ni):
    Lat = np.zeros(ni)
    wgt = np.zeros(ni)
    Tgg = 0.0
    swg = 0
    cni= ni
    pi= 3.141592
    for i in range(ni):
        Lat[i] = -85 + (i-1)*10
        wgt[i] = math.cos(abs(Lat[i])*pi/180.)
        swg   = swg + wgt[i] 
        Tgg = Tgg + Temp[i]*wgt[i]
    Tgg = Tgg/swg
    return Tgg

TG = [] # This list will store T global for ploting later
TL = [] # This list will store T lat at equilibrium for ploting later
Time = [] # This list will store  the time for T global plot

for j in range(500):
    Tg = Calculate_Tglobal(Tlat,Tgg,18)
    for i in range(18):
        Rin     = sfrac[i]*SR*(1-albedo[i])  # Incoming Energy
        Rout    = 4.0*ems*sigma*Tlat[i]**4   # Outgoing Enegy
        Trans   = tc*(Tlat[i] - Tg)          # Energy Flux
        Fn      = (Rin - Rout - Trans)/thc   # Tendency
        Tf      = Tlat[i] + Fn*dt               
        Tlat[i] = Tf
        if(j==499):
            TL.append(Tlat[i]-273.14)
    TG.append(Tg-273.14)
    Time.append(j*dt)

plt.plot(Time,TG)
plt.show()
zone = [-90,-80,-70,-60,-50,-40,-30,-20,-10,10,20,30,40,50,60,70,80,90]
```
